```python
target_cols = ['<HIGH>','<LOW>', '<OPEN>', '<CLOSE>']
df_brent_new = df_brent.groupby(['<DATE>'])[target_cols].agg({'<HIGH>': max,'<LOW>': min ,'<OPEN>': lambda x: x.iloc[0], '<CLOSE>': lambda x: x.iloc[-1]}).reset_index()
df_brent_new.head()
```
