## Ввод данных и дальнейшее приведение колонок с датами к datetime

```python
df_brent = pd.read_csv('brentrub.csv', 'r', delimiter=';')
df_money = pd.read_csv('eur_usd_test.csv', 'r', delimiter=';')
```

```python
df_brent['<DATE>'] = pd.to_datetime(df_brent['<DATE>'], format='%d/%m/%y')
df_brent['<TIME>'] = pd.to_datetime(df_brent['<TIME>'], format='%H:%M:%S').dt.time
df_money['DATE'] = pd.to_datetime(df_money['DATE'], format='%d/%m/%y')
df_brent = df_brent.dropna()
```

```python
max_per_days = df_brent[['<DATE>', '<HIGH>']].groupby('<DATE>', as_index=False).max()
min_per_days = df_brent[['<DATE>', '<LOW>']].groupby('<DATE>', as_index=False).min()
```

```python
max_per_days.columns = ['DATE', 'HIGH BRENT, $']
min_per_days.columns = ['DATE', 'LOW BRENT, $']
```

```python
df_money1 = pd.merge(df_money[['DATE', 'EUR/RUB', 'USD/RUB']], max_per_days, on=('DATE'), how='left')
df_money2 = pd.merge(df_money1, min_per_days, on=('DATE'), how='left')
df_money2
```

```python
open_per_days = df_brent[['<DATE>', '<OPEN>']].groupby('<DATE>', as_index=False).agg('first')
close_per_days = df_brent[['<DATE>', '<CLOSE>']].groupby('<DATE>', as_index=False).agg('last')
open_per_days.columns = ['DATE', 'OPEN BRENT, $']
close_per_days.columns = ['DATE', 'CLOSE BRENT, $']
```

```python
df_money3 = pd.merge(df_money2, open_per_days, on=('DATE'), how='left')
df_money_brent = pd.merge(df_money3, close_per_days, on=('DATE'), how='left')
```

```python
df_money_brent.head(7)
```

```python
df_money_brent.to_csv('money_brent.csv', sep=';')
```
